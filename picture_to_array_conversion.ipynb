# Intro to ImageIO and Translating Image Files to Array

```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 9)
```

```python
im = imageio.imread('imageio:chelsea.png')
print(im.shape)
```

```python
im_gray = np.dot(im[...,:3], [.3, .6, .1])
print(im_gray.shape)
```

```python
plt.imshow(im)
```

```python
plt.imshow(im_gray, cmap='gray')
```
